### Chapter 3, Section 6: Derivatives of Logarithmic and Inverse Trigonometric Functions
- Differentiation rules for exponents and bases are summarized on page 221
- Problems: Paper (3, 5, 7, 9, 13, 21, 25, 45,51, 55, 63, 65, 69, 71).
- Problems: Jupyter (41).

```python
import numpy as np
import matplotlib.pyplot as plt

# Choose domain (avoid x <= 0)
x_min = 0.05
x_max = 20
x = np.linspace(x_min, x_max, 2000)

# Functions
f = np.sin(x) + np.log(x)
fprime = np.cos(x) + 1.0/x

# Create plot
plt.figure(figsize=(10,6))
plt.plot(x, f, label=r'$f(x)=\sin x + \ln x$', color='C0', linewidth=2)
plt.plot(x, fprime, label=r"$f'(x)=\cos x + \dfrac{1}{x}$", color='C1', linewidth=2, linestyle='--')

# Highlight near-zero behavior region (optional)
plt.axvline(1.0, color='gray', linewidth=0.6, linestyle=':')

# Labels and legend
plt.title('Function and Derivative')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(alpha=0.3)
plt.legend()
plt.xlim(x_min, x_max)

# Optionally set y-limits to see both clearly; comment out if you prefer autoscale
# plt.ylim(-5, 5)

plt.show()

```
