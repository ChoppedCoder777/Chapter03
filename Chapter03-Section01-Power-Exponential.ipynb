```python



import numpy as np
import matplotlib.pyplot as plt

# Define the function and its derivative
def f(x):
    return 3*x**2 - x**3

def f_prime(x):
    return 6*x - 3*x**2  # derivative of f(x)

# Point of tangency
x0 = 1
y0 = f(x0)
slope = f_prime(x0)

# Tangent line: y = m(x - x0) + y0
def tangent_line(x):
    return slope * (x - x0) + y0

# Explicit tangent line equation in slope-intercept form
b = y0 - slope * x0
tangent_eq = f"y = {slope}x + {b}"

# Print the tangent line equation in the console
print("Tangent line equation:", tangent_eq)

# Create x values for plotting
x_vals = np.linspace(-1, 4, 400)
y_vals = f(x_vals)
tangent_vals = tangent_line(x_vals)

# Plotting
plt.figure(figsize=(8, 6))
plt.plot(x_vals, y_vals, label='Curve: y = 3x - x', color='blue')
plt.plot(x_vals, tangent_vals, label=f'Tangent at (1, 2): {tangent_eq}', 
         color='red', linestyle='--')
plt.scatter(x0, y0, color='black', zorder=5)
plt.text(x0 + 0.1, y0 + 0.5, '(1, 2)', fontsize=12)

plt.title('Curve and Tangent Line  Homework 3.1 #43')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()

# This line is essential in VS Code or when running as a script
plt.show()
```

```python



import numpy as np
import matplotlib.pyplot as plt

# Define f(x) and f'(x)
def f(x):
    return x**4 - 2*x**3 + x - 1

def f_prime(x):
    return 4*x**3 - 6*x**2 + 1

# Create x values
x_vals = np.linspace(-2, 3, 400)
y_vals = f(x_vals)
y_prime_vals = f_prime(x_vals)

# Plotting
plt.figure(figsize=(8, 6))
plt.plot(x_vals, y_vals, label='f(x) = x^4 - 2x^3 + x - 1', color='blue')
plt.plot(x_vals, y_prime_vals, label="f'(x) = 4x^3 - 6x^2 + 1", color='red', linestyle='--')

plt.axhline(0, color='black', linewidth=0.8)  # x-axis
plt.axvline(0, color='black', linewidth=0.8)  # y-axis

plt.title("Function and Derivative Comparison")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()
plt.show()
```

```python

import numpy as np
import matplotlib.pyplot as plt

# Define f(x), f'(x), and f''(x)
def f(x):
    return 2*x - (5/4)*x**3

def f_prime(x):
    return 2 - (15/4)*x**2

def f_double_prime(x):
    return -(15/2)*x

# Domain for plotting
x = np.linspace(-3, 3, 400)

# Evaluate
y = f(x)
y1 = f_prime(x)
y2 = f_double_prime(x)

# Plot all three together
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='f(x) = 2x - (5/4)x', color='blue')
plt.plot(x, y1, label="f'(x) = 2 - (15/4)x", color='red', linestyle='--')
plt.plot(x, y2, label="f''(x) = -(15/2)x", color='green', linestyle=':')

plt.axhline(0, color='black', linewidth=0.8)
plt.axvline(0, color='black', linewidth=0.8)

plt.title("Function, First Derivative, and Second Derivative")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()
plt.show()
```
