```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#graph of problem 35 on page 189.
x0 = np.linspace(0, 4, 400)
fx = (x0**2)/(1+x0)
x1 = np.linspace(-1, 3, 400)
tangent = (3/4)*x1 - 1/4
normal = -(4/3)*x1 + 11/6

# Plot 
plt.plot(x0, fx, label=r'${\frac{x^2}{1 + x}}$', color='black')
plt.plot(x1, tangent, label=r'${tangent}$', color='red')
plt.plot(x1, normal, label=r'${normal}$', color='green')
plt.title(' ')
plt.xlabel('x')
plt.axis('equal')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return (3*x) / (1 + 5*(x**2))

x = np.linspace(-4, 4, 1000)
y = f(x)

x0, y0 = 1, 0.5

# Tangent and normal
m_tan = -1/3
y_tan = m_tan*(x - x0) + y0

m_norm = 3
y_norm = m_norm*(x - x0) + y0

plt.figure(figsize=(8,5))
plt.plot(x, y, color="black", linewidth=2, label="y = 3x / (1+5x^2)")
plt.plot(x, y_tan, color="red", linewidth=2, label="Tangent line")
plt.plot(x, y_norm, color="green", linewidth=2, label="Normal line")
plt.scatter([x0], [y0], color="red", s=60, zorder=3, label="(1, 1/2)")

plt.axhline(0, color="black", linewidth=1)
plt.axvline(0, color="black", linewidth=1)

plt.grid(True, linestyle="--", alpha=0.4)
plt.xlim(-4, 4)
plt.ylim(-3, 3)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Curve with Tangent and Normal at (1, 1/2)")
plt.legend(frameon=False, loc="upper left")

plt.tight_layout()
plt.show()
```
